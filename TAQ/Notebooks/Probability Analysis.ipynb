# Emperical Probability of Up Move given Create/Join

```python
#Get CJ Flagger df

import sys
sys.path.insert(0, "../Helper_Input")
from iex_helper import * #imports stock_data class from iex_helper.py
import time
#######################################################
start = time.time()
quotes = '../Training_Files/AAPL_train.csv' #
q = pd.read_csv(quotes,low_memory = False)
AAPL_QW = Quote_Wrangler(quotes) #initiate class
from IPython.display import display
from IPython.display import Markdown
end = time.time()
print('Time to Completion: ', end-start, " seconds")

NBBO = AAPL_QW.NB_master
```

Now, using NBO_all or NBB_all, we find the next unique mid, and save this as its own column. The functionality for this is found as seperate helper functions in iex_helper

```python
NBBO = get_next_mid(NBBO_cj_Master)
NBBO.head(15)
```

Compute Difference column, as this will be usefull in the next steps


```python
NBBO['Mid_Change'] = NBBO.Mid_Next - NBBO.Mid
#This is the last step before function two cells below
```

Now, we want to splice this by creates/joins AND by exchange. We start with an example - Exchange T Creates

```python
#First Isolate Exchange
P_flag = NBBO.B_Exchanges.apply(lambda x :'P' in x)
BBO_P = NBBO.copy()
BBO_P['Ex_Flag'] = P_flag
BBO_P = BBO_P[BBO_P.Ex_Flag == True]
display(BBO_P.head(20))
print('\n')
#Next Isolate Creates/Joins
BBO_P_Create = BBO_P[BBO_P.Creates != '']
BBO_P_Join = BBO_P[BBO_P.Joins != '']
display(Markdown('BBO P Exchange, Joins'))
display(BBO_P_Join.head())
```

Now that we have isolated exchange P, Joins, we can find the empirical probability of an up move

```python
up_count = sum(np.array(BBO_P_Join.Mid_Change > 0))
prob = up_count/len(BBO_P_Join)
print("Probability of an Up Move: ", prob)
```

Using this method, we can iterate though all exchanges for Creates or Joins and find the emperical prob


```python
exch = list(AAPL_QW.exchange_map.keys())
def prob_finder(NBX,exch):
    """
    NBX should be NBBO with cj and mid_next
    """
    prob_matrix = np.empty([len(exch),4])
    NBX_BB = NBX[NBX.Flag == 'NBB']
    NBX_BO = NBX[NBX.Flag == 'NBO']
    for i in range(len(exch)):
        ex = exch[i]
        NBX_BB_ex = NBX_BB.copy()
        NBX_BO_ex = NBX_BO.copy()
        
        NBX_BB_ex['Ex_Flag'] =  NBX_BB_ex.B_Exchanges.apply(lambda x :ex in x)
        NBX_BB_ex = NBX_BB_ex[NBX_BB_ex.Ex_Flag == True]

        NBX_BO_ex['Ex_Flag'] =  NBX_BO_ex.A_Exchanges.apply(lambda x :ex in x)
        NBX_BO_ex = NBX_BO_ex[NBX_BO_ex.Ex_Flag == True]
            
        NBX_BB_ex_C = NBX_BB_ex[NBX_BB_ex.Creates != '']
        NBX_BB_ex_J = NBX_BB_ex[NBX_BB_ex.Joins != '']

        NBX_BO_ex_C = NBX_BO_ex[NBX_BO_ex.Creates != '']
        NBX_BO_ex_J = NBX_BO_ex[NBX_BO_ex.Joins != '']
        
        try:
            prob_BB_c = sum(np.array(NBX_BB_ex_C.Mid_Change > 0))/len(NBX_BB_ex_C)
        except:
            prob_BB_c = 0
        try:
            prob_BB_j = sum(np.array(NBX_BB_ex_J.Mid_Change > 0))/len(NBX_BB_ex_J)
        except:
            prob_BB_j = 0
        try:
            prob_BO_c = sum(np.array(NBX_BO_ex_C.Mid_Change > 0))/len(NBX_BO_ex_C)
        except:
            prob_BO_c = 0
        try:
            prob_BO_j = sum(np.array(NBX_BO_ex_J.Mid_Change > 0))/len(NBX_BO_ex_J)
        except:
            prob_BO_j = 0
            
            
        prob_matrix[i,0] = prob_BB_c
        prob_matrix[i,1] = prob_BB_j
        prob_matrix[i,2] = prob_BO_c
        prob_matrix[i,3] = prob_BO_j
    prob_df = pd.DataFrame(prob_matrix)
    prob_df.index = exch
    prob_df.columns = ['Best_Bid_Creates','Best_Bid_Joins','Best_Offer_Creates','Best_Offer_Joins']
    return prob_df

NBBO_probs = prob_finder(NBBO,exch)

# NBO_probs[(NBO_probs.Creates > 0) & (NBO_probs.Joins > 0)]
```
